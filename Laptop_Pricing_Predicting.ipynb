```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Task 1: Load the dataset
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
df = pd.read_csv(filepath, header=None)
```

```python
#print the head to preview the dataset
print(df.head())
```

```python
#Task 2: Add headers to the dataframe
#The headers for the dataset, in sequence, are "Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg" and "Price".
#Confirm insertion by printing the first 10 rows of the dataset.
headers = [['Manufacturer', 'Category', 'Screen', 'GPU', 'OS', 'CPU_core', 'Screen_Size_cm', 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'Weight_kg', 'Price']]
df.columns = headers
df.head(10)
```

```python
#Task 3: Replace the '?' entries in the dataset with NaN value, recevied from the Numpy package.
df.replace("?", np.nan, inplace=True)
```

```python
#Task 4: Print the data types of the dataframe columns
df.dtypes
```

```python
#Task 5: Print the statistical description of the dataset, including that of 'object' data types.
df.describe(include = 'all').transpose()
```

```python
#Task 6: check missing values in the dataset
df.isnull().sum()
```

```python
#change the data type of screen size and weight kg to float data type.
df[['Weight_kg','Screen_Size_cm']] = df[['Weight_kg','Screen_Size_cm']].astype('float').round(2)
df
```

```python
# replacing missing data with mean
avg_weight=df['Weight_kg'].mean()
df['Weight_kg'] = df['Weight_kg'].replace(np.nan, avg_weight)
df
```

```python
common_screen_size = df['Screen_Size_cm'].mode().iloc[0]
df['Screen_Size_cm'] = df['Screen_Size_cm'].fillna(common_screen_size)
df.isnull().sum()
```

```python
#Data Standaradization
# Data standardization: convert weight from kg to pounds
df["Weight_kg"] = df["Weight_kg"]*2.205
df.rename(columns={'Weight_kg':'Weight_pounds'}, inplace=True)

# Data standardization: convert screen size from cm to inch
df["Screen_Size_cm"] = df["Screen_Size_cm"]/2.54
df.rename(columns={'Screen_Size_cm':'Screen_Size_inch'}, inplace=True)
```

```python
df.head(5)
```

```python
# Data Normalization
df['CPU_frequency'] = df['CPU_frequency']/df['CPU_frequency'].max()
df.head()
```

```python
df['Price'] = df['Price'].astype(int)
```

```python
# Data Binnings
print(type(df['Price']))
print(df['Price'].shape)
```

```python
price_series = df['Price'].squeeze()  # Converts the DataFrame to a Series
print(type(price_series))
print(price_series.shape)
```

```python
# Define bins and labels
bins = np.linspace(price_series.min(), price_series.max(), 4)
group_names = ['Low', 'Medium', 'High']

# Bin the data
df['Price-binned'] = pd.cut(price_series, bins, labels=group_names, include_lowest=True)

# Check the DataFrame
df
```

```python
plt.bar(group_names, df["Price-binned"].value_counts())
plt.xlabel("Price")
plt.ylabel("count")
plt.title("Price bins")
```

```python
# Variable indicator
# Flatten column names if they are tuples
df.columns = [''.join(col) if isinstance(col, tuple) else col for col in df.columns]
```

```python
# Verify column names after flattening
print("Columns in DataFrame after flattening:")
print(df.columns)
```

```python
# Create dummy variables from the 'Screen' column
dummy_variable_1 = pd.get_dummies(df['Screen'], prefix='Screen')

# Print dummy variables DataFrame to verify
print("Dummy variables DataFrame:")
print(dummy_variable_1)
```

```python
# Concatenate dummy variables with the original DataFrame
df = pd.concat([df, dummy_variable_1], axis=1)

# Drop the original 'Screen' column
df.drop('Screen', axis=1, inplace=True)

# Print the final DataFrame
df
```

```python
df.to_csv("Cleanedlaptop_df.csv")
```
